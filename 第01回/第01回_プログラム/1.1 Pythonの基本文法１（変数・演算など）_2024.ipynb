# 数値や文字列の画面表示
## print関数
- 文字列や数値などを画面に出力する関数<br>
`print(表示内容)`
  - 文字列の表示："abc"（ダブルクォーテーション）または'abc'（シングルクォーテーション）で括る
  - 数値の表示：ダブルクォーテーションやシングルクォーテーションで括る必要なし
  - `print関数`で表示される文字列や数値の最後に改行（改行コード）が入る

**「Hello World」という文字列を表示**

```python
print("Hello World")    # 文字列はダブルクォーテーションで括る
```

**#はコメント（メモ書き）を表し、#以降の実行されない**

**数値の表示**

```python
print(10)
print(5.5)
```

## 複数の値のprint関数
- 複数の値を表示する場合は`,`(カンマ)区切りで続けて記述
- print関数では最後に自動で改行が入る（次にprint関数を使った場合は行が変わる）

```python
print(10, "円")
print("時給",1000,"円")
```

## 最後に改行で入れない場合
- print(値, end="改行コードの代わりに用いる文字や記号")
- 複数のprint関数の結果が1行で表示される

```python
print(10, end=" ")
print(5.5, end=" ")
print("ok")
```

# 変数と型
## 変数とは
- 変数は値が出し入れする箱のようなもので、`1つ`の値だけを入れることができる
- 変数に出し入れする値とは`数値`や`文字列`など
- 変数に値を入れることを`代入`という（特に、初めて値を入れる場合は`初期化`という）
- 変数の値は何度でも入れ替えることができる（ただし、古い値は消える）
<br><br>
- 値の性質によって`データ型`という分類があり、`データ型`によって計算機能は異なる
  - 特に、`整数`、`浮動小数点数（小数）` 、`文字`を扱う型は`基本データ型`と呼ばれる
  - 代入される値の種類によって変数には型が設定される
    - 整数: `int型` (例: 3, 10, 5)
    - 浮動小数点数（小数）: `float型` (例: 3.14)
    - 文字: `str型` (例: ABC)
    - Pythonは、変数に対して型の記述は不要（他言語では型を宣言して記述する必要があるものもある）。ただし、変数の型を意識する必要はある。例えば、「3」が整数か文字なのかで、四則演算できるかできないかに影響する。


<br><br>
- 変数は以下の例外を除いて自由に名前を付けることができる
  - 変数名の一文字目には数字は使えない
  - +, -, /, %, (, ), # のような演算子記号は使えない
  - Pythonで予め用意されている予約語(and, as, breakなど)や、組み込みオブジェクト(while、if、sumなど)は使えない
  - 変数名はその変数がどのような意味を持つかが分かるように命名することが重要

## 変数の初期化（変数に初めて値を入れる）
- ファイルを閉じなければ、変数の値は以降のセルにも保持される

```python
num1  = 123             # 整数
num2  = 123.456         # 小数
text  = "Hello World!"  # 文字列
```

**プログラムの式は、左辺←右辺の操作を意味する。算数や数学の等式とは意味合いが異なることに注意。**

## 変数の画面表示
- 変数にどのような値が代入されているかは、`print関数`で表示できる

**各変数の値を表示**

```python
print(num1)
print(num2)
print(text)
```

**print関数ではコンマで区切って変数を並べることもできる**<br>
表示はスペース区切り

```python
print(num1, num2, text)
```

```python
num1  = 987            # 整数
num2  = 567.23         # 小数
text  = "See you."     # 文字列

print("num1:", num1)
print("num2:", num2)
print("text:", text)
```

# 算術演算子
プログラム上で演算を行うためには、算術演算子を用いる

||||
|:-:|:-:|:--|
|算術演算子|\+  | 足し算 |
|          | -  | 引き算 |
|          | *  | かける |
|          | /  | 割る（小数） |
|          | // | 割る（整数） |
|          | %  | 余り |
|          | ** | べき乗 |
||||

## 算術演算

```python
# 足し算
print("足し算", 10 + 3)

# 引き算
print("引き算", 10 - 3)

# 掛け算
print("掛け算", 10 * 3)

# 割り算
print("割り算", 10 / 3)

# 割り算の商
print("割り算の商", 10 // 3)

# 割り算の余り
print("割り算の余り", 10 % 3)

# べき乗
print("3乗", 10 ** 3)
```

## 変数を使った算術演算
- 変数を使って演算式を書くと、変数の値を書き変えるだけで全ての演算結果が得られる

```python
num1 = 10
num2 = 3

# 足し算
print("足し算", num1 + num2)

# 引き算
print("引き算", num1 - num2)

# 掛け算
print("掛け算", num1 * num2)

# 割り算
print("割り算", num1 / num2)

# 割り算の商
print("割り算の商", num1 // num2)

# 割り算の余り
print("割り算の余り", num1 % num2)

# べき乗
print("3乗", num1 ** num2)
```

## 演算結果を新たな変数に代入して保持

```python
num1 = 10
num2 = 3

# 足し算
tasu = num1 + num2
print("足し算", tasu)

# 引き算
hiku = num1 - num2
print("引き算", hiku)

# 掛け算
kake = num1 * num2 
print("掛け算", kake)

# 割り算
waru = num1 / num2
print("割り算", waru)

# 割り算の商
shou = num1 // num2
print("割り算の余り", shou)

# 割り算の余り
amari = num1 % num2
print("割り算の余り", amari)

# べき乗
beki = num1 ** num2
print("3乗", beki)
```

## 複数の変数の演算
- 直方体の体積を計算するプログラム

```python
edge1 = 10   # 辺1(cm)
edge2 = 20   # 辺2(cm)
edge3 = 30   # 高さ(cm)

volume = edge1*edge2*edge3 # 体積の計算

print("体積：", volume, "cm3")
```
