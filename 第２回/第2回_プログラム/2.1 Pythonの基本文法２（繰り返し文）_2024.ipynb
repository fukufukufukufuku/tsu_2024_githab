# 制御構文
- プログラムに高度な機能（電卓以上）を持たせるためには制御構文が必要
- 制御構文とは、プログラムの実行順序を制御するためにもの
- プログラムは、`順次`，`選択（分岐）`，`反復（繰り返し、ループ)`の3つ制御構文の組み合わせである
- この3つによって、あらゆる実行順序が記述できる



----

# 反復（繰り返し、ループ）
- 反復（繰り返し、ループ）には`for文`や`while文`を用いる
- `for文`は、指定した回数を繰り返す（繰り返し範囲を指定するためには、リストやrangeをin演算子とともに用いる）
- `while文`は、ある条件を満たすまで繰り返す
- 繰り返しの途中で処理を中断したい場合には、`break文`を使う
- 繰り返しの途中で特定の処理だスキップしたい場合には、`continue文`を使う

**今回は、特に重要なfor文を扱います。**

# for文を用いた反復（繰り返し、ループ）
繰り返し範囲を指定するためには、`range関数`や`リスト`のオブジェクト（データの集合）をin演算子とともに用いる。<br>
データを1個ずつ変数に代入して反復処理を行う
```
for 変数 in オブジェクト:
    実行する処理
```

**[注意事項]**<br>
- for文の処理を記述する際は`空白（半角スペース）を入れて字下げ`しなければならない（「`インデントを揃える`」という）。通常、半角スペース4つ
- 毎回半角スペースのキーを4回打つのは面倒なので、`Tab`キーを押せば自動的に調整してくれる
- **リストについては、次回の学習予定のためrange関数のみを扱う**

## `range関数`の使い方
```
range(n)
```
- **`0`から`n-1`までの`n個`の整数を順番に並べたデータ集合を生成**
- 例えば、range(3)の場合、{0,1,2}のデータの集合が生成される<br>
- for文では0,1,2を順番に変数に代入して処理を繰り返す(つまり、3回処理を反復)
<br>
```
range(開始番号, 終了番号)
```
- 開始番号から開始して、 `終了番号-1`までの整数を順番に並べたデータ集合を生成
<br>
```
range(開始番号, 終了番号, ステップ数)
```
- 開始番号から開始して、 `終了番号-1`まで、ステップ数の間隔で整数を順番に並べたデータ集合を生成

## range関数によりN回繰り返す

**0から9までの整数をカンマ区切りで表示するプログラム**

````python
for i in range(10):
    print(i, end=',')
````

**「programming languate」をカンマ区切りで10回表示するプログラム**

````python
for i in range(10):
    print("programming language", end=',')
````

## range関数により特定の範囲で繰り返す

**5から9までの整数をカンマ区切りで表示するプログラム**

````python
for i in range(5, 10):  # 終了番号は1加える
    print(i, end=',')
````

## range関数により特定の間隔で繰り返す

**0から9までの偶数をカンマ区切りで表示するプログラム**

````python
for i in range(0, 10, 2):  # 終了番号は1加える
    print(i, end=',')
````

# for文を使った演算

## 整数の合計値を計算

**1から100までの`整数`を合計するプログラム**

````python
total = 0                 # 合計を保持する変数で、初期値を0としておく

for i in range(1, 101):  # 終了番号は1加える 
    total = total + i    # totalにiを加算

print("合計: ", total)    # 合計出力
````

**1から100までの`偶数`を合計するプログラム**

````python
total = 0                   # 合計を保持する変数で、初期値を0としておく

for i in range(2, 101, 2): # 開始番号は2, 終了番号は1加える 
    total = total + i      # totalにiを加算
    
print("合計: ", total)     # 合計出力
````

## 練習

(1) 1から10000までの奇数の合計を計算してみよう。合計を表す変数は`total`とする。最後に合計を出力してみよう。

(2) 初期投資が1000円で、毎日の利率0.1%で運用した場合、365日後にはいくらになるかを計算してみよう（365回の繰り返し）。初期投資額を表す変数は`initial`、合計を表す変数は`total`とする。最後に初期投資からの利益を出力してみよう。

# for文とif文の組み合わせ

**1から1000までの整数の中で、7の倍数以外の合計を計算するプログラム**

````python
total = 0                      # 合計を保存する変数で、初期値を0としておく
for i in range(1, 1001):      # 終了番号は1加える 
    amari = i%7               # あまり計算
    if amari != 0:            # 条件
        total = total + i     # totalにiを加算
    
print("合計: ", total)        # 合計出力
````

**8月に31日間アルバイトをしたとする。1日目から15日目までは日給は3000円、16日目から31日目までは3500円であった場合、8月の合計収入を計算してみよう。for文でカレンダー上の1日から31日まで反復とする。**

````python
total = 0                      # 合計を保存する変数で、初期値を0としておく

for i in range(1, 32):         # カレンダーの1日から31日まで繰り返し(終了番号に1加える)
    if i <= 15:
        total = total + 3000   # 3000を加算
    else:
        total = total + 3500   # 3500を加算  

print("合計: ", total, "円")   # 合計出力
````

## 練習

(1) 1から500までの整数の中で、5の倍数以外の合計を計算してみよう。合計を表す変数は`total`とする。最後に合計を出力してみよう。

(2) 8月に31日間アルバイトをしたとする。偶数の日の日給は4100円、奇数の日の日給は3800円であった場合、8月の合計収入を計算してみよう。for文でカレンダー上の1日から31日まで反復とする。偶数と奇数の判定は、2で割った余りが0か0以外かで判定できる。合計を表す変数は`total`とする。最後に合計を出力してみよう。
