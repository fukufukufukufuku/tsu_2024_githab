# プログラミングにおける関数の概要

## 関数とは
- ある値を入力すると一定の処理がされて結果が返ってくる機能のこと
- 使用頻度が高い処理を纏めるために関数が利用できる


## 関数のメリット
- 機能がまとめられているとプログラム全体がわかりやすい
- 同じ処理を何度も書く必要がなくなる
- 入力と出力の関係を理解していれば、複雑な関数の中身を意識しなくてもよい

## 関数を使う手順
- (1) 関数を設定
  ```Python
  # 関数の記述方法の基本
  def 関数名(引数):
      処理
      retrun 戻り値
  ```
  - 関数名は機能がわかる名前を付ける
  - `関数の処理内容は字下げして記述（インデントを下げる, 半角スペース4個分）`
  - `入力`に対応する引数は関数名に続けて()内で記述
  - 引数が複数ある場合はコンマ区切りで並べる`(引数1, 引数2,....)`  
  - 関数を設定する際の引数には変数を用いる（仮引数と呼ぶ）
  - 引数がない場合は()のみを記述し、引数は記述不要
  - `出力`に対応する戻り値は`retrun`で返す
  - 戻り値がない場合は記述不要  
<br><br>  
- (2) 定義した関数を実行
  ```Python
  # 関数の使い方の基本
  変数名 = 関数名(引数)
  ```
  - `入力`に対応する引数は関数名に続けて()内で記述
  - 関数を使用する場合の引数には具体的な値を入れる（実引数と呼ぶ）
  - 引数がない場合は()のみを記述し、引数は記述不要
  - `出力`に対応する戻り値は`変数名`で受け取る
  - 戻り値がない場合は、`関数名(引数)`のみ記述

  
## 組み込み関数
- Pythonには言語が標準で持っている`組み込み関数`が多く用意されている
- 文字列を出力する`print文`は代表的な組み込み関数
  ```Python
  print(引数)
  ```

# 関数の基本
2つの値を割り算する関数を作成
- 関数名：`division`(divisionは割り算の意味)
- 引数: `x1`, `y1`
- 結果を代入する変数：`rslt`(rsltはresultの略)
- 戻り値：`rslt`
- 関数を使って計算する割り算
  - 1回目：100÷40
  - 2回目：50÷30

## 戻り値あり

````python
# 関数定義
def division(x1, y1):    # 引数: x1, y1
    rslt = x1 / y1       # 処理内容の記述
    return rslt         # 戻り値: rslt
````

````python
# 関数呼び出し（1回目）
rslt = division(100, 40) # 関数に渡す値を引数に記述
print("割り算", rslt)    # 結果表示

# 関数呼び出し（2回目）
rslt = division(50, 30)  # 関数に渡す値を引数に記述
print("割り算", rslt)    # 結果表示
````

## 戻り値なし

````python
# 関数定義
def division(x1, y1):     # 引数: x1, y1
    rslt = x1 / y1        # 処理内容の記述
    print("割り算", rslt) # 関数内で結果表示
````

````python
# 関数呼び出し（1回目）
division(100, 40)         # 関数に渡す値を引数に記述

# 関数呼び出し（2回目）
division(50, 30)          # 関数に渡す値を引数に記述
````

# if文を含む関数
試験の点数から合格を判定する関数を作成
- 関数名：`pass_judgement`
- 引数: `score`
- 判定条件: 60点以上：`合格`、60点未満：`不合格`
- 結果を代入する変数：`rslt`(rsltはresultの略)
- 戻り値: `rslt`
- 関数を使って判定する学生
  - 学生1: 72点
  - 学生2: 65点
  - 学生3: 50点

````python
# 関数定義
def pass_judgement(score): # 引数:score
    if score > 60:     
        rslt = "合格"
    else:
        rslt = "不合格"        
    return rslt            # 戻り値:rslt
````

````python
# 関数呼び出し（1回目）
rslt = pass_judgement(72)  # 関数に渡す値を引数に記述
print("判定1", rslt)       # 結果表示

# 関数呼び出し（2回目）
rslt = pass_judgement(65)  # 関数に渡す値を引数に記述
print("判定2", rslt)       # 結果表示

# 関数呼び出し（3回目）
rslt = pass_judgement(50)  # 関数に渡す値を引数に記述
print("判定3", rslt)       # 結果表示
````

# for文を含む関数
開始の整数（start）から終わりの整数(end)までを足し合わせる関数を作成してみよう。
- 関数名：`summation`（summationは合計の意味）
- 引数: `start`, `end`
- 結果を代入する変数：`rslt`(rsltはresultの略)
- 戻り値: なし
- 関数を使って計算する整数の範囲
  - 1回目：1から10
  - 2回目：21から57
  - 3回目：104から219

````python
# 関数定義
def summation(start, end):          # 引数: start, end
    rslt = 0                        # 初期値を0に設定
    for i in range(start, end+1):  # for文で計算(end+1までであることに注意)
        rslt = rslt + i
    print("計算結果", rslt)         # 結果表示
````

````python
# 関数呼び出し（1回目）
summation(1,10)             # 関数に渡す値を引数に記述

# 関数呼び出し（2回目）
summation(21,57)            # 関数に渡す値を引数に記述

# 関数呼び出し（3回目）
summation(104,219)          # 関数に渡す値を引数に記述
````
