# 比較演算子
- 等しい(==)、〇〇以上(>=)、〇〇未満(<)などのように左辺と右辺の比較に使う演算記号を、`比較演算子`と言う
- 出力は`True`（真）または`False`（偽）
- 左辺の変数に右辺の値を代入する場合は`=`であったが、左辺と右辺の値を比較する場合は`==`を使うことに注意

||||
|:-:|:-:|:--|
|比較演算子| <  | 小さい     |
|          | >  | 大きい     |
|          | <= | 以上       |
|          | >= | 以下       |
|          | == | 等しい     |
|          | != | 等しくない |
||||

## 数値の比較

````python
num1 = 10

# 等しい
print("判定1: ", num1 == 10)

# 等しくない
print("判定2: ", num1 != 10)

# 大きい
print("判定3: ", num1 > 10)

# 小さい
print("判定4: ", num1 < 10)

# 以上
print("判定5: ", num1 >= 10)

# 以下
print("判定6: ", num1 <= 10)
````

# 論理演算子
- 2つ以上の比較条件の組み合わせを判定する場合に使う
- 出力は`True`（真）または`False`（偽）

||||
|:-:|:-:|:--|
|論理演算子| and | 両者を満たす |
|          | or | どちらか片方を満たす |
|          | not | 満たさない |
||||




## ２つの比較条件の組み合わせ

````python
num1 = 10
num2 = 20

# 条件1と条件2の両方を満たすか
print("判定1: ", num1 >= 10 and num2 <= 10)

# 条件1と条件2のどちらかを満たすか
print("判定2: ", num1 >= 10 or num2 <= 10)

# 条件を満たさないか
print("判定3: ", not num1 >= 10)

# 条件を満たさないか
print("判定4: ", not num2 <= 10)
````

## ３つの比較条件の組み合わせ

````python
num1 = 10
num2 = 20
num3 = 30

# 条件1と条件2と条件3を同時に満たすか
print("判定1: ", num1 >= 10 and num2 <= 10 and num3 == 10)

# 条件1と条件2と条件3のいずれかを満たすか
print("判定2: ", num1 >= 10 or num2 <= 10 or num3 == 10)
````

# 制御構文
- プログラムに高度な機能（電卓以上）を持たせるためには制御構文が必要
- 制御構文とは、プログラムの実行順序を制御するためにもの
- プログラムは、`順次`，`選択（分岐）`，`反復（繰り返し、ループ)`の3つ制御構文の組み合わせである
- この3つによって、あらゆる実行順序が記述できる

# 選択（分岐）
## 条件文（if文）
- 選択（分岐）には`条件文（if文）`を用いる
- if文を用いることで、**「もし〇〇だったら、処理A。そうでなく△△だったら、処理B。」**のように、条件によって処理を変更できる
<br><br>

- **1つの条件を判定する場合**
```Python
if 条件式1:
    条件式1がTrue(真)ならば、ここに記述された処理を行う
```
条件式の後に`:`(コロン)が付く（**付け忘れしやすいので注意**）

- **複数の条件を判定する場合**
```Python
if 条件式1:
    条件式1がTrue(真)ならば、ここに記述された処理を行う
elif 条件式2:
    条件式1がFalse(偽)で、条件式2がTrue(真)ならば、ここに記述された処理を行う
else:
    全ての条件式がFalse(偽)ならば、ここに記述された処理を行う
```

ifの条件式が満たされていなければ、elifの条件式が上から順番に判定される<br>
これらの条件式がすべて満たされていなければ、else内の処理が実行される<br>



**[注意事項]**<br>
- if文の処理を記述する際は`空白（半角スペース）を入れて字下げ`しなければならない（「`インデントを揃える`」という）。通常、半角スペース4つ
- 毎回半角スペースのキーを4回打つのは面倒なので、`Tab`キーを押せば自動的に調整してくれる

## 条件が1個
**変数numが100未満なら「AAA」と出力するプログラム**

````python
num = 50
if num < 100:
    print("AAA")
````

## 条件が2個
**変数numが100未満なら「AAA」、それ以外は「BBB」と出力するプログラム**

````python
num = 101
if num < 100:
    print("AAA")
else:
    print("BBB")
````

## 条件が3個
**変数numが100未満なら「AAA」、100以上200未満なら「BBB」、それ以外は「CCC」と出力するプログラム**

````python
num = 150

if num < 100:
    print("AAA")
elif num < 200:
    print("BBB")
else:
    print("CCC")
````

## 条件が4個
**変数numが100未満なら「AAA」、100以上200未満なら「BBB」、200以上300未満なら「CCC」、それ以外は「DDD」と出力するプログラム**

````python
num = 150

if num < 100:
    print("AAA")
elif num < 200:
    print("BBB")
elif num < 300:
    print("CCC")    
else:
    print("DDD")
````

## 論理演算子を用いたif文

**トランプカードがダイアで10以上ならポイントが10点、それ以外なら2点加算されるプログラムを作成する。初期値として100ポイントを持つ**

````python
points = 100 # ポイント、初期値

num = 10          # トランプの番号
mark = "diamond"  # トランプのマーク

if num >=10 and mark == "diamond":
    points = points + 10
else:
    points = points + 2
    
print(points)
````
