# Googleの画像検索で検索された画像の取得
- AIへの質問例：Googleの画像検索で得られた学生の画像10枚をフォルダに保存するプログラム。フォルダの場所はdrive/MyDrive/Colab Notebooks/第05回_プログラム/student_imagesとする。

```python
import cv2
import glob
from google.colab.patches import cv2_imshow
!pip install icrawler
from icrawler.builtin import GoogleImageCrawler
import os

# Googleドライブをマウント
from google.colab import drive
drive.mount('/content/drive')

# 保存先のフォルダパスを入力
save_dir = 'drive/MyDrive/Colab Notebooks/第05回_プログラム/student_images'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

google_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=2,
    downloader_threads=4,
    storage={'root_dir': save_dir})

# 検索キーワードと枚数を指定してクロール
google_crawler.crawl(keyword='学生', max_num=10)

print(f"画像が{save_dir}に保存されました。")
```

# 写真から人の顔を検出して表示
- AIへの質問例：フォルダ内の写真から人の顔を検出して表示するプログラム。画像フォルダは/content/drive/MyDrive/Colab Notebooks/第05回_プログラム/student_imagesとする。

```python
# prompt: フォルダ内の写真から人の顔を検出して表示するプログラム。画像フォルダは/content/drive/MyDrive/Colab Notebooks/第05回_プログラム/student_imagesとする。

import cv2
import glob
from google.colab.patches import cv2_imshow

# 顔検出のための学習済みモデルを読み込み
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 画像が保存されているフォルダパス
image_dir = 'drive/MyDrive/Colab Notebooks/第05回_プログラム/student_images'

# フォルダ内の画像ファイルをリストアップ
image_files = glob.glob(image_dir + '/*.jpg')

# 各画像に対して顔検出と枠付けを行う
for image_file in image_files:
    # 画像を読み込み
    img = cv2.imread(image_file)

    # グレースケールに変換
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 顔検出
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # 検出された顔に枠付け
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # 枠付けされた画像を表示
    cv2_imshow(img)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```
